```python
# Data cleaning - 1
# Fixing rows and columns
# There are lots of columns with null values(entire column). So, we can remove those columns to reduce the data size to be analyzed.
loan_csv.dropna(inplace=True, axis=1, how='all')
loan_csv
```

```python
# Data cleaning - 2
# Now after the step of Data cleaning - 1, the number of column size has gradually reduced to 57. There are still some columns which
# can be eliminated as they don't contribute much to the case study analysis.
loan_csv.drop(['pymnt_plan','initial_list_status','collections_12_mths_ex_med','policy_code','acc_now_delinq', 'application_type', 'pub_rec_bankruptcies', 'tax_liens', 'delinq_amnt'],axis=1,inplace=True) 
```

```python
# Now after stage 2 of Data cleaning, there are 48 columns left in the loan_data.csv 
# Out of these 48 columns, we need to further clean/remove the columns which are not required for the loan lending case study.
# The following columns can be removed because of the mentioned reasons:
# "id","member_id","url","title","emp_title","zip_code","addr_state","last_credit_pull_d" and "desc" can be removed as these columns
# have irrelevant information for the loan defaulting prediciton/analysis
# "funded_amnt" column can also be removed as "funded_amnt_inv" is the actual amount funded and "funded_amnt" is just the estimation.
# "delinq_2yrs","revol_bal","out_prncp","total_pymnt","total_rec_prncp","total_rec_int","total_rec_late_fee","recoveries","collection_recovery_fee"
# "last_pymnt_d","last_pymnt_amnt","next_pymnt_d","chargeoff_within_12_mnths","mnths_since_last_delinq","mnths_since_last_record" are 
# the columns which can be removed as it has info on post loan approval features and doesn't contribute much to the pre-loan lending process.

loan_csv.drop(["id", "member_id", "url", "title", "emp_title", "zip_code", "last_credit_pull_d", "addr_state","desc","out_prncp_inv","total_pymnt_inv","funded_amnt", "delinq_2yrs", "revol_bal", "out_prncp", "total_pymnt", "total_rec_prncp", "total_rec_int", "total_rec_late_fee", "recoveries", "collection_recovery_fee", "last_pymnt_d", "last_pymnt_amnt", "next_pymnt_d" , "chargeoff_within_12_mths", "mths_since_last_delinq", "mths_since_last_record"], axis = 1, inplace = True)
```

```python
# Now out of these 21 columns "loan_status" columns is the main and it can also be used to reduce the no of rows as "loan_status" value
# "Current" cannot be used further for this case study as it's in progress.
# Removing the row/records with loan_status as current

loan_csv = loan_csv[loan_csv["loan_status"] != "Current"]
loan_csv
```

```python
# Fixing missing values:
# Now, we have to find the missing row values and fill it with mean values
loan_csv.isna().sum()
```

```python
# Percentage of missing values
(loan_csv.isna().sum()/len(loan_csv.index))*100
```

```python
# So, there are two columns "emp_length" and "revol_util" with missing values. We need to fill those missing values with appropriate values
loan_csv.info()
```

```python
# So from the above info we can see the data type of "emp_length" and "revol_util" is object
# It's better to find the mode of emp_length and see whether the other data/values is nearer to mode
loan_csv.emp_length.mode()
```

```python
# Find the frequency of data in emp_length columns
loan_csv.emp_length.value_counts()
```

```python
# As from the above data, we clearly see the mode value can be filled for the missing values as it is having far higher frequency than
# other values
mode=loan_csv.emp_length.mode()
loan_csv.emp_length.fillna(mode[0], inplace=True)
loan_csv.emp_length.isna().sum()
```

```python
# For revol_util columns, since it's very small % of data is empty, we can use dropna to drop the rows
loan_csv.dropna(axis=0, subset = ['revol_util'] , inplace = True)
loan_csv.revol_util.isna().sum()
```

```python
# Standardizing the values
# As fixing rows and columns, fixing missing values is done, now we have to standardize the data/value mainly standardizing the formats,
# removing outliners, standardising the units, standardising the precisions, standardising the text etc.,

loan_csv.head()
```

```python
# Firstly emp_length column is inconsistent with values 10+ years and <1year, so we can convert < 1 year as 0 and 10+ years as 10
loan_csv["emp_length"] = loan_csv["emp_length"].astype("string")
loan_csv.emp_length = pd.to_numeric(loan_csv.emp_length.apply(lambda x: 0 if "<" in x else (x.split('+')[0] if "+" in x else x.split()[0])))
loan_csv.head()
```

```python
# Also we need to remove % from the int_rate column
loan_csv["int_rate"] = loan_csv["int_rate"].astype("string")
loan_csv.int_rate = pd.to_numeric(loan_csv.int_rate.apply(lambda x: x.split('%')[0]))
loan_csv.head()
```

```python
# Also we need to remove % from revol_util
loan_csv.revol_util = loan_csv.revol_util.astype("string")
loan_csv.revol_util = pd.to_numeric(loan_csv.revol_util.apply(lambda x: x.split('%')[0]))
loan_csv.head()
```

```python
# Derived Metrics - Creating new columns with bins which will help for bi-variate analysis
# creating bins for int_rate,open_acc,revol_util,total_acc
loan_csv['int_rate_bin'] = pd.cut(loan_csv['int_rate'], bins=5,precision =0,labels=['5%-10%','10%-15%','15%-19%','19%-21%','21%-24%'])
loan_csv['open_acc_bin'] = pd.cut(loan_csv['open_acc'],bins = 5,precision =0,labels=['2-10','10-20','20-25','25-35','35-44'])
loan_csv['revol_util_bin'] = pd.cut(loan_csv['revol_util'], bins=5,precision =0,labels=['0-20','20-40','40-60','60-80','80-100'])
loan_csv['total_acc_bin'] = pd.cut(loan_csv['total_acc'], bins=5,precision =0,labels=['2-20','20-35','35-55','55-75','75-90'])
loan_csv['annual_inc_bin'] = pd.cut(loan_csv['annual_inc'], bins=5,precision =0,labels =['3k-31k','31k-58k','58k-85k','85k-112k','112k-140k'])
```

```python
# BiVariate Analysis
# Now we can analyze the bins created with the actual columns
# Let's first start with interest rate bin vs interest rate
fig, ax = plt.subplots(figsize = (15,10))
plt.subplot(221)
sns.countplot(x='int_rate_bin', data=loan_csv[loan_csv.loan_status == 'Charged Off'])
```

```python
# Obs-1: So, the analysis from the above plot is that there are more people whose fall under the interest rate 15-19
# So, there is more probability of defaulting when the interest_rate falls under 15-19%
```
